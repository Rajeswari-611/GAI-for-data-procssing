```python
import torch
from diffusers import PixArtAlphaPipeline

model_id = "PixArt-alpha/PixArt-XL-2-512x512"

dtype = torch.float16 if torch.cuda.is_available() else torch.float32

pipe = PixArtAlphaPipeline.from_pretrained(
    model_id,
    torch_dtype=dtype,
)

device = "cuda" if torch.cuda.is_available() else "cpu"

pipe = pipe.to(device)

# Optional: enable faster attention if available
if device == "cuda":
    pipe.enable_model_cpu_offload()  # works well on limited VRAM
    pipe.enable_vae_slicing()
```

```python
prompt = (
    "Professional headshot of a person in formal attire, natural smile, "
    "white background, high detail, sharp focus, realistic skin texture, "
    "studio-quality lighting, portrait photography"
)

negative_prompt = (
    "blurry, low quality, deformed face, asymmetry, extra limbs, distorted features, "
    "harsh shadows, watermark, text, logo"
)
```

```python
generator = torch.Generator(device=device).manual_seed(42) # set seed for reproducibility

image = pipe (

prompt=prompt,

negative_prompt=negative_prompt,

num_inference_steps=24, # increase for slightly higher quality

guidance_scale=4.5, #3.5-6 is typical

height=512,
width=512,

generator=generator

).images[0]

#4) Save the result

image.save("face.png")

print("Saved to face.png")
```
